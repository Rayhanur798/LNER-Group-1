```python
#merging datasets
```

```python
delay = delay.rename(columns={'Rep. Period': 'Period'})
delay['Date'] = pd.to_datetime(delay['Date'], errors='coerce')
delay['Period'] = delay['Period'].astype(str)
delay['Period'] = delay['Period'].str[-2:]
delay['Period'] = delay['Period'].str.zfill(2)
delay['Period'] = delay['Date'].dt.strftime('%y') + delay['Period']

performance['Period'] = performance['Period'].astype(str).str.strip()

delay_period = delay.groupby('Period', as_index=False).agg({
    'Delay Mins': 'sum',
    'Tot. Mins': 'sum',
    'Inc.#': 'count'
})
delay_period = delay_period.rename(columns={'Inc.#': 'Delay Incident Count'})

performance_period = performance.groupby('Period', as_index=False).agg({
    'Total Mins': 'sum',
    'Incident Number': 'count'
})
performance_period = performance_period.rename(columns={'Incident Number': 'Performance Incident Count'})

master_df = pd.merge(delay_period, performance_period, on='Period', how='outer')
master_df = master_df.sort_values('Period')

print(master_df.shape)
print(master_df.head(10))
print(master_df.tail(10))

master_df.to_csv('/content/master_fleet_dataset.csv', index=False)
```
